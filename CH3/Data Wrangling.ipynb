# Data Wrangling


Data wrangling is a term often used to describe the early stages of the data analytics process. It involves transforming and mapping data from one format into another. The aim is to make data more accessible for things like business analytics or machine learning. 


Data wrangling is a broad term used, often informally, to describe the process of
transforming raw data to a clean and organized format ready for use. For us, data
wrangling is only one step in preprocessing our data, but it is an important step.

The most common data structure used to “wrangle” data is the data frame, which can
be both intuitive and incredibly versatile. Data frames are tabular, meaning that they
are based on rows and columns like you would see in a spreadsheet

```python
import pandas as pd 
```

```python
dataframe = pd.DataFrame()
```

```python
dataframe
```

```python
dataframe['Name'] = ['Mohammad' , 'Ahmed' , 'motafa' , 'Fares']
dataframe['Age'] = [12,15,60,20]
dataframe['Gender'] = ['Male' , 'Male' , 'Male' , 'Male']
dataframe['Location'] = ['Cairo' , 'London' , 'US' , 'KSA']
```

## Concatenating DataFrames

### Problem
You want to concatenate two DataFrames.

<h3 style ="color : green">Solution ? <h3>

```python
data_a = {'id': ['1', '2', '3'],
 'first': ['Alex', 'Amy', 'Allen'],
 'last': ['Anderson', 'Ackerman', 'Ali']}
```

```python
dataframe_1 = pd.DataFrame(data_a , columns= ['id', 'first', 'last'])
```

```python
dataframe_1
```

```python
data_b = {'id': ['4', '5', '6'],
 'first': ['Billy', 'Brian', 'Bran'],
 'last': ['Bonder', 'Black', 'Balwner']}
```

```python
dataframe_2 = pd.DataFrame(data_b , columns= ['id', 'first', 'last'])
```

```python
dataframe_2
```

```python
dataframe_3 = pd.concat([dataframe_1 , dataframe_2] , axis=0)
```

```python
dataframe_3
```
